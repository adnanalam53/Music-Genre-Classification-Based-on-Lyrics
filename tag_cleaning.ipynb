```
import re
import operator

ly = pd.read_csv("new_lyrics_with_tags.csv", dtype = {'lyrics':str}) #read the cleanned file with 77885 entries
flag = 0
for i in range(len(ly)):
  flag += 1
  #a = []
  tag = re.split(r'\W', ly["tags"][i].lower())
  # create a dictionary to check the occurances of each genre
  genre = {"rock": tag.count("rock"),
          "pop" : tag.count("pop"),
          "alternative" : tag.count("alternative"),
          "indie" : tag.count("indie"),
          "electronic" : tag.count("electronic"),
          "female_vocalists" : tag.count("female vocalists"),
          "favorites" : tag.count("favorites"),
          "dance" : tag.count("dance"),
          "00s" : tag.count("00s")}
  ly["tags"][i] = max(genre.items(), key=operator.itemgetter(1))[0] #get the label with highest occurances from a single song
  if(flag == 2000):
    print(i, ly["tags"][i])
    flag = 0
  


  
  


```

```
rank = {"rock":len(ly[ly["tags"]=="rock"]),
       "pop":len(ly[ly["tags"]=="pop"]),
       "alternative":len(ly[ly["tags"]=="alternative"]),
       "indie":len(ly[ly["tags"]=="indie"]),
       "electronic":len(ly[ly["tags"]=="electronic"]),
       "female vocalists":len(ly[ly["tags"]=="female vocalists"]),
       "favorites":len(ly[ly["tags"]=="favorites"]),
       "love":len(ly[ly["tags"]=="love"]),
       "dance":len(ly[ly["tags"]=="dance"]),
       "00s":len(ly[ly["tags"]=="00s"])}
#rank
sorted_rank = sorted(rank.items(), key=operator.itemgetter(1),reverse=True)
sorted_rank
#So the 5 most occurred genres by order are: "rock", "pop", "indie", "alternative", "dance"
```
